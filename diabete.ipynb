```python
import csv

header = ["notp", "pga", "dip","tsft","2hoursi","bmi","def","age","Class"]

with open('../input/testdiabete/diabete.csv', 'r') as fp:
    reader = csv.DictReader(fp, fieldnames=header)


    with open('diabete.csv', 'w', newline='') as fh: 
        writer = csv.DictWriter(fh, fieldnames=reader.fieldnames)
        writer.writeheader()
        header_mapping = next(reader)
        writer.writerows(reader)
```

1. notp: Hamile sayısı
2. pga: Oral glukoz tolerans testinde 2 saat plazma glikoz konsantrasyonu
3. dip: Diyastolik kan basıncı (mm Hg)
4. tsft: Triceps deri kıvrım kalınlığı (mm)
5. 2hoursi: 2 saatlik serum insülini (mu U / ml)
6. bmi: Vücut kitle indeksi (kg olarak ağırlık / (m olarak yükseklik) ^ 2)
7. def: Diyabet soyağacı fonksiyonu
8. age: Yaş
9. Class: Sınıf değişkeni
